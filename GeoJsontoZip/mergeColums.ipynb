***This Notebook is Intended to Merge data from the Demographics CSV and the Solar Panel Detection***

```python
import pandas as pd
```

```python
df_unemployment = pd.read_csv("/home/elsherif/Desktop/Thesis/ViewPython/CSV_data/Unemployment/cleaned_districts_unemployment.csv")  
df_tiles = pd.read_csv("/home/elsherif/Desktop/Thesis/ViewPython/GeoJsontoZip/tiles_with_district_year.csv")
df_average_age = pd.read_csv("/home/elsherif/Desktop/Thesis/ViewPython/CSV_data/population/average_age_population.csv")
df_old = pd.read_csv("/home/elsherif/Desktop/Thesis/ViewPython/CSV_data/population/old_population.csv")
df_young = pd.read_csv("/home/elsherif/Desktop/Thesis/ViewPython/CSV_data/population/young_population.csv")
df_total_population = pd.read_csv("/home/elsherif/Desktop/Thesis/ViewPython/CSV_data/population/overall_population.csv")
df_houses = pd.read_csv("/home/elsherif/Desktop/Thesis/ViewPython/CSV_data/Houses/standAloneHouses.csv")
```

```python
df_unemployment["District"] = df_unemployment["District"].astype(str).str.zfill(2)
df_average_age["District"] = df_average_age["District"].astype(str).str.zfill(2)
df_old["District"] = df_old["District"].astype(str).str.zfill(2)
df_young["District"] = df_young["District"].astype(str).str.zfill(2)
df_total_population["District"] = df_total_population["District"].astype(str).str.zfill(2)
df_houses["District"] = df_houses["District"].astype(str).str.zfill(2)
df_tiles["district_number"] = df_tiles["district_number"].astype(str).str.zfill(2)
```

```python
df_merged = df_tiles.merge(
    df_unemployment[['Year', 'District', 'Value']],
    left_on=['year', 'district_number'],
    right_on=['Year', 'District'],
    how='left'
)
```

```python
df_merged = df_merged.drop(columns=['Year', 'District'])
```

```python
df_merged = df_merged.rename(columns={"Value": "Unemployment_Rate"})
```

```python
df_merged = df_merged.drop(columns=["tile_centroid_lat", "tile_centroid_lon", "district_name"])
```

```python
df_merged = df_merged.merge(
    df_average_age[['Year', 'District', 'Population']],
    left_on=['year', 'district_number'],
    right_on=['Year', 'District'],
    how='left'
)
```

```python
df_merged = df_merged.merge(
    df_young[['Year', 'District', 'Population']],
    left_on=['year', 'district_number'],
    right_on=['Year', 'District'],
    how='left'
)
```

```python
df_merged = df_merged.drop(columns=['Year', 'District'])
df_merged = df_merged.rename(columns={"Population": "Young_Population"})
```

```python
df_merged = df_merged.drop(columns=['Year', 'District'])
df_merged = df_merged.rename(columns={"Overall Population": "Total_Population"})
```
