Step 1 : Merge all Rooftop CSV files

```python
import pandas as pd
import glob

# Path to your folder containing the CSVs
path = "/home/elsherif/Desktop/Thesis/ViewPython/data/Rooftop_Data/clean/*.csv"

# Read all CSVs
csv_files = glob.glob(path)

dfs = []

for file in csv_files:
    df = pd.read_csv(file)
    print(f"Loaded {file} with shape {df.shape}")
    # Optional: Add filename or year from filename
    # Example: extract year like Munich2018 → 2018
    # import re
    # year = re.findall(r"\d{4}", file)[0]
    # df["year"] = int(year)
    
    dfs.append(df)

# Merge vertically (stack rows)
merged_df = pd.concat(dfs, ignore_index=True)

# Save result
merged_df.to_csv("/home/elsherif/Desktop/Thesis/ViewPython/data/Rooftop_Data/clean/MergedRooftop.csv", index=False)

print("Merged into MergedRooftop.csv")
```

```python
df_tiles = pd.read_csv("/home/elsherif/Desktop/Thesis/ViewPython/data/Rooftop_Data/clean/MergedRooftop.csv")
df_unemployment = pd.read_csv("/home/elsherif/Desktop/Thesis/ViewPython/data/CSV_data/Unemployment/cleaned_districts_unemployment.csv")  
df_average_age = pd.read_csv("/home/elsherif/Desktop/Thesis/ViewPython/data/CSV_data/population/average_age_population.csv")
df_old = pd.read_csv("/home/elsherif/Desktop/Thesis/ViewPython/data/CSV_data/population/old_population.csv")
df_young = pd.read_csv("/home/elsherif/Desktop/Thesis/ViewPython/data/CSV_data/population/young_population.csv")
df_total_population = pd.read_csv("/home/elsherif/Desktop/Thesis/ViewPython/data/CSV_data/population/overall_population.csv")
df_houses = pd.read_csv("/home/elsherif/Desktop/Thesis/ViewPython/data/CSV_data/Houses/standAloneHouses.csv")
```

```python
df_unemployment["District"] = df_unemployment["District"].astype(str).str.zfill(2)
df_average_age["District"] = df_average_age["District"].astype(str).str.zfill(2)
df_old["District"] = df_old["District"].astype(str).str.zfill(2)
df_young["District"] = df_young["District"].astype(str).str.zfill(2)
df_total_population["District"] = df_total_population["District"].astype(str).str.zfill(2)
df_houses["District"] = df_houses["District"].astype(str).str.zfill(2)
df_tiles["district_number"] = df_tiles["district_number"].astype(str).str.zfill(2)
```
