**Analysis and Processing of Population Data**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
population_data = pd.read_excel("population_munich.xlsx" , sheet_name=1)
```

```python
population_data.head()
```

```python
population_data['Indikator'].unique()
```

```python
unique_pairs = population_data[['Indikator', 'Ausprägung']].drop_duplicates()
print(unique_pairs)
```

**Overall Polulation**

```python
df_overallpopulation = population_data[(population_data['Indikator'] == 'Altersgruppen') & (population_data['Ausprägung'] == 'insgesamt')]
```

```python
df_overallpopulation.head()
```

```python
df_overallpopulation = df_overallpopulation.drop(columns=['Indikator','Basiswert 1', 'Basiswert 2', 'Basiswert 3', 'Basiswert 4', 'Basiswert 5', 'Name Basiswert 1','Name Basiswert 2', 'Name Basiswert 3', 'Name Basiswert 4', 'Name Basiswert 5','Ausprägung'])
```

```python
df_overallpopulation.rename(columns={'Indikatorwert': 'Overall Population', "Raumbezug": "Location" , "Jahr": "Year"}, inplace=True)
```

```python
df_overallpopulation.head()
```

```python
df_overallpopulation = df_overallpopulation.reset_index(drop=True)
df_overallpopulation['Year'] = df_overallpopulation['Year'].astype(int)
```

```python
df_overallpopulation['Location'].unique()
```

```python

# Work on a copy
df = df_overallpopulation.copy()

# Ensure correct column names
df.columns = [c.strip() for c in df.columns]

# Convert columns to numeric
df["Overall Population"] = pd.to_numeric(df["Overall Population"], errors="coerce")
df["Year"] = pd.to_numeric(df["Year"], errors="coerce")

# Create pivot: rows = Year, columns = Location, values = Overall Population
pivot_df = df.pivot(index="Year", columns="Location", values="Overall Population")

# Choose locations to exclude from the legend (optional)
exclude = ["Stadt München"]  # example: remove the overall city line from legend

# Plot
plt.figure(figsize=(16, 8))  # Adjust figure size

for col in pivot_df.columns:
    if col not in exclude:
        plt.plot(
            pivot_df.index,
            pivot_df[col],
            label=col,
            linewidth=2,       # line thickness
            linestyle='-',     # line style
            marker='o',        # marker at each data point
            alpha=0.8          # transparency
        )

plt.ylabel("Population", fontsize=14)
plt.xlabel("Year", fontsize=14)
plt.title("Population Change Over Time per District in Munich", fontsize=16)

# Customize legend
plt.legend(
    title="Location",
    bbox_to_anchor=(1.05, 1),  # move outside to the right
    loc="upper left",
    fontsize=10,
    title_fontsize=12
)

plt.tight_layout()  # adjust layout to fit everything
plt.show()
```

```python

df_overallpopulation['District'] = df_overallpopulation['Location'].str.extract(r'^(\d{2})')  # Extract the first two digits

# Handle rows like 'Stadt München' (no digits)
df_overallpopulation['District'] = df_overallpopulation['District'].fillna('00')

# Remove the leading number and dash from Location text
df_overallpopulation['Location'] = df_overallpopulation['Location'].str.replace(r'^\d{2}\s*-\s*', '', regex=True)
df_overallpopulation['Location'] = df_overallpopulation['Location'].str.replace(r'^\d{2}\s*', '', regex=True)
# (Optional) Clean up whitespace
df_overallpopulation['Location'] = df_overallpopulation['Location'].str.strip()

print(df_overallpopulation.head(10))
```

```python
# Save your DataFrame as a CSV file
df_overallpopulation.to_csv("overall_population.csv", index=False, encoding='utf-8')
```
